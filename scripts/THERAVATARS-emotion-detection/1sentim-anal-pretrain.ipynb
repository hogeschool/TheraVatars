STEP 2: <br>
-We noticed that the pre-trained sentiment classifier is not working for the neutral classes.<br>
-So we remove the neutral class, and we will only classify for negative and positive classes<br>
-Only consider 500 randomly selected positive, and 500 randomly selected negative samples (use random state) for classification.<br>
-Map the labels to numbers.

-We will remove the class of neutrals<br>
-We will make a smaller data set of ONLY 500 positive and 500 negative labels, randomly selected, using a random_state = 42.<br>
-Random_state, randomizes your data, but always in the same way. This makes comparison between several model versions possible.

```python
###STUDENTS###
"""
Remove the class of 'neutral',
if you know that next code line only keeps positive sentiments, adapt it to remove 'neutral' and keep only 'negative' and 'positive' sentiment classes.
df = df[df.airline_sentiment == 'negative']
df = df[df['airline_sentiment'] == 'negative']
complete code below:
"""
#df = df[df.airline_sentiment != 'neutral']
df = df[df['airline_sentiment']!='neutral']
```

```python
df
```

```python
###STUDENTS : complete code###

# Sample random rows with 500 negative class labels, 500 positive class labels
negative_indices = df[df['airline_sentiment'] == 'negative'].sample(n=500, random_state=42).index
positive_indices = df[df['airline_sentiment'] == 'positive'].sample(n=500, random_state=42).index

# Create separate DataFrames for positive and selected negative instances, use loc
df_positive = df.loc[positive_indices]
df_negative = df.loc[negative_indices]

# Concatenate the DataFrames to combine them
df = pd.concat([df_positive, df_negative])
```

```python
###STUDENTS###
df['airline_sentiment'].value_counts()
```

```python
df.head()
```

map positive labels to 1 and negative labels to 0

```python
###STUDENTS : complete dictionary target_map###
target_map = {'positive' : 1, 'negative' : 0}
df['target'] = df['airline_sentiment'].map(target_map)
```

##6 Metrics

STEP 4:<br>
-Apply the metrics below in order to compare your output predictions (preds) and your reference target labels df['target']

```python
print('accuracy', np.mean(df['target']==preds))
```

```python
# Compute F1 score, precision, and recall for the general result and per predicted class
report = classification_report(df['target'], preds, target_names=['negative', 'positive'])
print(report)
```

```python
cm = confusion_matrix(df['target'], preds, normalize='true')
#normaliz. sums up to 1
cm
```

```python
def plot_cm(cm):
    classes = ['negative','positive']
    df_cm = pd.DataFrame(cm, index=classes, columns=classes)
    ax = sn.heatmap(df_cm, annot=True, fmt='g')
    ax.set_xlabel("Predicted")
    ax.set_ylabel("Target")


plot_cm(cm)
```

###6.1 Confidence measures positive and negative predictions

A Confidence Score is a number between 0 and 1 that represents the <br>
likelihood that the output of a Machine Learning model is correct.<br>
On top of accuracy and F score, it can express the confidence with which the model predicts if 2 models are compared, and have a similar accuracy or F score.<br> An analysis of the confidence measure can support a good evaluation of the models.

```python
pos_scores = [p['score'] for p in predictions if p['label'].startswith('P')]
neg_scores = [p['score'] for p in predictions if p['label'].startswith('N')]

pos_scores[:10]
```

```python
neg_scores[:10]
```

```python
avg_pos_conf=np.mean(pos_scores)
avg_neg_conf=np.mean(neg_scores)
```
